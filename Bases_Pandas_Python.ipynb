#Ejemplo 5: MAnipulación de columnas

```python
import pandas as pd
```

```python
datos = {
        "Nombre": ["Pepinillo Hernández", "Lúpulo de Dios", "Juan Juon", "Jimmy el Patatas", "Lorenzo Retaguardias"],
        "Cereal favorito": ["Korn Floks", "Verdurinis", "Zumbaritas", "Diabetukis, Papá", "Fibra Máxima 3000"],
        "Hora del desayuno": ["11:00", "07:30", "07:00", "08:30", "09:30"]
    }

df = pd.DataFrame(datos)
df
```

```python
#Agregar columnas al DataFrame
df["Fruta con la que acompaña el cereal"] = pd.Series(['Pera', 'Manzana', 'Plátano', 'Guayaba', 'Pizza'])
df
```

```python
#Asignar una nueva serie a una columna que ya existe (modificar valores de una columna)
df["Hora del desayuno"] = pd.Series(['10:30', '6:30', '8:00', '7:45', '11:20'])
df
```

```python
#Eliminar una columna de datos del dataframe
df_dropped= df.drop(columns=['Fruta con la que acompaña el cereal'])
df_dropped
```

#Ejemplo 6: Lectura de JSON

```python
import pandas as pd
import json

#load : a través de una ruta, toma un archivo en modo lectura y lo convierte en un diccionario
#loads : Espera una cadena de caracteres (no un archivo) pero con estructura de JSON y lo convierte en un diccionario

#dump: Convierte un diccionario a un JSON
#dumps : Convierte un diccionario a cadena de caracteres

#Abrimos el dataset llamado Zomato reviews en modo lectura por la 'r'
f = open('/content/drive/MyDrive/Módulo 2 Ciencia de datos/Datasets/zomato_reviews-clean.json','r')
```

```python
#Abrimos el archivo en formato json para convertirlo en un diccionario
json_data = json.load(f)
#Siempre se debe cerrar un archivo que hayamos abierto así como una conexión a base de datos por seguridad de los procesos que estemos usando
f.close()
```

```python
df = pd.DataFrame.from_dict(json_data)
df
```
